```python
i = 0
v = '27'

print(v)
print('------------------')
while True:
    p = int(v[i]) * int(v[i+1])
    v = v + str(p)
    print(v)
    print('6 count: {}'.format(v.count('6')))
    print('6 %: {}'.format(v.count('6')/len(v)))
    print('------------------')
    i += 1
    
    if i == 100:
        break
```
