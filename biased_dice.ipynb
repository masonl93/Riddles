```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
```

```python
def monte_carlo_simulation(num_simulations, bias_probability=1/6):
    count_double_six = 0
    count_three_sixes = 0
    count_four_sixes = 0
    count_five_sixes = 0
    
    # Calculate the probabilities for the other numbers
    other_probability = (1 - bias_probability) / 5
    probabilities = [other_probability] * 5 + [bias_probability]
    
    for _ in range(num_simulations):
        rolls = np.random.choice([1, 2, 3, 4, 5, 6], size=5, p=probabilities)
        
        # Check for consecutive sixes
        if rolls[0] == 6 and rolls[1] == 6:
            count_double_six += 1
            if rolls[2] == 6:
                count_three_sixes += 1
                if rolls[3] == 6:
                    count_four_sixes += 1
                    if rolls[4] == 6:
                        count_five_sixes += 1
    
    return {
        "double_six": (count_double_six / num_simulations) * 100,
        "three_sixes": (count_three_sixes / num_simulations) * 100,
        "four_sixes": (count_four_sixes / num_simulations) * 100,
        "five_sixes": (count_five_sixes / num_simulations) * 100
    }

# Calculate how much more likely biased results are compared to unbiased results
def calculate_likelihood_ratio(biased, unbiased):
    return biased / unbiased if unbiased != 0 else float('inf')
```

## 2/7 Chance of rolling a 6


```python
num_simulations = 100000

# Unbiased dice
unbiased_results = monte_carlo_simulation(num_simulations, bias_probability=1/6)
print(f"Percentage of double sixes with unbiased dice: {unbiased_results['double_six']:.2f}%")
print(f"Percentage of three sixes with unbiased dice: {unbiased_results['three_sixes']:.2f}%")
print(f"Percentage of four sixes with unbiased dice: {unbiased_results['four_sixes']:.2f}%")
print(f"Percentage of five sixes with unbiased dice: {unbiased_results['five_sixes']:.2f}%")
print()

# Biased dice
biased_results = monte_carlo_simulation(num_simulations, bias_probability=2/7)
print(f"Percentage of double sixes with biased dice: {biased_results['double_six']:.2f}%")
print(f"Percentage of three sixes with biased dice: {biased_results['three_sixes']:.2f}%")
print(f"Percentage of four sixes with biased dice: {biased_results['four_sixes']:.2f}%")
print(f"Percentage of five sixes with biased dice: {biased_results['five_sixes']:.2f}%")
print()

likelihood_ratios = {
    "double_six": calculate_likelihood_ratio(biased_results['double_six'], unbiased_results['double_six']),
    "three_sixes": calculate_likelihood_ratio(biased_results['three_sixes'], unbiased_results['three_sixes']),
    "four_sixes": calculate_likelihood_ratio(biased_results['four_sixes'], unbiased_results['four_sixes']),
    "five_sixes": calculate_likelihood_ratio(biased_results['five_sixes'], unbiased_results['five_sixes'])
}

print(f"Likelihood ratio of double sixes (biased vs unbiased): {likelihood_ratios['double_six']:.2f}")
print(f"Likelihood ratio of three sixes (biased vs unbiased): {likelihood_ratios['three_sixes']:.2f}")
print(f"Likelihood ratio of four sixes (biased vs unbiased): {likelihood_ratios['four_sixes']:.2f}")
print(f"Likelihood ratio of five sixes (biased vs unbiased): {likelihood_ratios['five_sixes']:.2f}")
```

### Heart Surgeon Example
A heart surgeon has performed 60 heart transplants with 0 deaths. What would estimate their error rate to be?

$p_m = 1 - I^{-1}_{1/2}(n-m, m+1)$

Where:
- $p_m$ is the Maximum Ignorance Probability
- $I.(.,.)$ is the beta regularized function
- $n$ is the total number of trials
- $m$ is the number of successes


```python
def calculate_mip(n, m):
    return 1 - stats.beta.ppf(0.5, n - m, m + 1)
```

```python
n = 60  # total number of trials
m = 0   # number of successes (in this case, 0 deaths)

p_m = calculate_mip(n, m)
print(f"Maximum Ignorance Probability (MIP) for the heart surgeon: {p_m:.4f}")
```

```python


n_values = range(60, 101)
mip_values = [calculate_mip(n, 0) for n in n_values]

plt.figure(figsize=(10, 6))
plt.plot(n_values, mip_values, marker='o')
plt.title('Maximum Ignorance Probability vs. Number of Trials')
plt.xlabel('Number of Trials (n)')
plt.ylabel('Maximum Ignorance Probability (MIP)')
plt.grid(True)
plt.show()

print(f"MIP for n=60: {mip_values[0]:.4f}")
print(f"MIP for n=100: {mip_values[-1]:.4f}")
```

## Biased Dice


```python
n_values = range(2, 11)
mip_values = []

for n in n_values:
    m = 0 # Number of non-6 rolls
    mip = calculate_mip(n, m)
    mip_values.append(mip)
    print(f"For n={n}, m={m}: Probability of not rolling a 6 = {mip:.4f}")

plt.figure(figsize=(10, 6))
plt.plot(n_values, mip_values, marker='o')
plt.axhline(y=1/6, color='r', linestyle='--', label='Threshold (1/6)')
plt.title('Probability of Not Rolling a 6 vs. Number of Rolls')
plt.xlabel('Number of Rolls (n)')
plt.ylabel('Probability of Not Rolling a 6 (MIP)')
plt.grid(True)
plt.legend()
plt.show()

print(f"\nProbability of not rolling a 6 for n=2: {mip_values[0]:.4f}")
print(f"Probability of not rolling a 6 for n=10: {mip_values[-1]:.4f}")


threshold = 1/6
for n, mip in zip(n_values, mip_values):
    if mip < threshold:
        print(f"MIP value becomes less than {threshold} when n = {n}")
        break
```

After 4 rolls of all 6s, we can declare the dice is biased.
