```python
import random

import matplotlib.pyplot as plt
import numpy as np
```

```python
benford_values = [np.log10(1 + 1 / digit) for digit in range(1, 10)]


def first_digit(num):
    while num >= 10:
        num //= 10
    return num


def plot_vs_benford(data):
    first_digits = [first_digit(num) for num in data]
    digit_counts = {i: first_digits.count(i) for i in range(1, 10)}
    theoretical = [perc * len(data) for perc in benford_values]

    plt.figure(figsize=(10, 6))
    digits = list(digit_counts.keys())
    counts = list(digit_counts.values())

    plt.bar(digits, counts, color="skyblue", label="Observed")

    plt.plot(
        digits, theoretical, marker="s", color="red", linestyle="--", label="Benford"
    )

    plt.title("Distribution of First Digits of Powers of 2 vs. Benford's Law")
    plt.xlabel("First Digit")
    plt.ylabel("Frequency")
    plt.xticks(digits)
    plt.legend()

    plt.grid(True, linestyle="--", alpha=0.7)
    plt.show()
```

# 2^n

```python
data = [2**n for n in range(1, 1001)]
plot_vs_benford(data)
```

# Uniform data

```python
data = [random.uniform(0, 1000000) for _ in range(1000)]
plot_vs_benford(data)
```

# Gaussian 

```python
data = np.random.normal(0, 1, 1000)
data = [x*100 for x in data]
plot_vs_benford(data)
```
